```python
import pandas as pd
import numpy as np
```

```python
titles = pd.read_csv("Data/title-akas-us-only.csv", low_memory=False)
```

```python
titles.head(10)
```

```python
basics = pd.read_csv('Data/title.basics.tsv.gz', sep='\t', low_memory=False)
```

```python
basics.head(10)
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us_titles = basics['tconst'].isin(titles['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
basics = basics.replace({'\\N':np.nan})
```

```python
basics.info()
```

```python
# drop rows with null vaules in the runtime and genres columns
basics = basics.dropna(subset=['runtimeMinutes', 'genres'])
```

```python
# verify rows are dropped
basics['genres'].isna().sum()
```

```python
# verify rows are dropped
basics['runtimeMinutes'].isna().sum()
```

```python
# Filter to only keep full length movies 
filtered_basics = (basics['titleType'].str.contains('movie'))
```

```python
# check for missing values in the start year column
basics['startYear'].isna().sum()
```

```python
# changing null values to nonsense number
basics['startYear'] = basics['startYear'].fillna(9999)
basics['startYear'].isna().sum()
```

```python
# convert startYear to float type
basics['startYear'] = basics['startYear'].astype(float)
basics['startYear'].dtype
```

```python
bf = basics[filtered_basics]
bf
```

```python
bf.info()
```

```python
# Filter to keep movies with startYears that are from 2000 to 2022
year_bf_filter = (bf['startYear'] >=2000) & (bf['startYear'] <= 2022)
f_bf = bf[year_bf_filter]
f_bf.head()
```

```python
# checking genres for how many categories and how many documentaries
f_bf['genres'].value_counts()
```

```python
filter_documentaries = ~f_bf['genres'].str.contains('Documentary')
```

```python
t_df = f_bf[filter_documentaries]
```

```python
# confirm all documentaries have been removed
t_df['genres'].value_counts()
```

```python
t_df.info()
```

```python
t_df.head()
```

```python
t_df.to_csv('Data/titles_basics.csv')
```
